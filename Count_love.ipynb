```
import pandas as pd
from tqdm.notebook import tqdm

from nltk import word_tokenize, pos_tag, ne_chunk
from konlpy.tag import Kkma
```

```
path = "/content/drive/MyDrive/Colab/Prob_and_Stats/Data/"
```

```
df_lyrics = pd.read_excel(path + "lyrics_2000_2018.xlsx")
```

```
kkma = Kkma()
```

```
#시작, 끝 연도 설정
start_year = 2000
end_year = 2018
```

```
df_lyrics.head()
```

```
#노래마다 둘러보며 "사랑"이라는 단어가 등장하는지 판단
cnt_love = 0
total_cnt = 0
all_nouns = []

for col_num in tqdm(list(i for i in range(start_year, end_year+1))):

  for row in tqdm(range(100)):#각 연도마다 Top 100
    nouns = []
    if type(df_lyrics[str(col_num)][row]) == str: #비어있을 경우 nan 값 (=float) return
      total_cnt += 1 #가사 추출된 노래의 수 count
      nouns = kkma.nouns(df_lyrics[str(col_num)][row])

      for noun in nouns:
        if noun == "사랑": 
          cnt_love += 1
          break #"사랑" 단어는 한 번만 세기
```
