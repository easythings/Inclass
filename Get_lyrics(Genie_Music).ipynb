```
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait

from bs4 import BeautifulSoup
from tqdm.notebook import tqdm

import pandas as pd
```

```
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless') #내부 창을 띄울 수 없으므로 설정
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
```

```
url = "https://www.genie.co.kr/chart/musicHistory?year="
```

```
driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
```

```
df_info = pd.DataFrame()
```

```
err_num = 0

for year in tqdm(range(2000, 2020)):
    lyrics = []
    err_flag = 0

    driver.get(url + str(year) + "&category=0")
    driver.implicitly_wait(10)
    
    len_ = len(driver.find_elements_by_class_name("link"))

    for idx in tqdm(range(len_)):
        driver.find_elements_by_class_name("link")[idx].find_element_by_tag_name("a").click()
        driver.implicitly_wait(10)
        
        try:
            lyrics.append(driver.find_element_by_id("pLyrics").find_element_by_tag_name("p").text)        
        except:
            print("Exception")
            lyrics.append("")
            err_num += 1
            
        driver.back()
        driver.implicitly_wait(10)
        
    #51~100위 페이지로 넘어감
    driver.find_element_by_css_selector('#body-content > div.songlist-box > div.page-nav.rank-page-nav > a:nth-child(2)').click()
        
    len_ = len(driver.find_elements_by_class_name("link"))

    for idx in tqdm(range(len_)):
        driver.find_elements_by_class_name("link")[idx].find_element_by_tag_name("a").click()
        driver.implicitly_wait(10)
        
        try:
            lyrics.append(driver.find_element_by_id("pLyrics").find_element_by_tag_name("p").text)        
        except:
            print("Exception")
            lyrics.append("")
            err_num += 1
            
        driver.back()
        driver.implicitly_wait(10)
        
    df_info[str(year)] = lyrics
```
